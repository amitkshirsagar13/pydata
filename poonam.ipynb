```python
### Defining Variable -> Variable is nothing but place holder!!!
husband='Amit'
wife="Poonam"
son="Amogh"
flats=["Saheel",'GangaAurum',"DNV"]
flatCount=len(flats)

family=[husband, wife, son]

x=10
y=5

```

```python

### Print Information to console -> Logging
print("--- Printing String")
print("WhoAmI")
print(10)
print("--- Printing Variable")
# print(husband)
# print(wife)
print(flatCount)
print("--- Printing Array")
print(family)


### Tasks:
## add x and y and put it in 'total' and then print it
total = x+y
print("---Printing total")
print(total)
```

```python

### Print all flats one by one -> Loop through flats and print each on seperate line.
print("--- Looping through Array using 'for' in array")
for flat in flats:
    print(flat)
print("--- Looping through Array using 'for' in range")
for index in range(0,len(family)):
    print(family[index])

```

```python

print("--- Looping 'for' in range to print designs")
hight=7
for indexi in range(0,hight):
    for indexj in range(0,hight):
        if indexi==indexj:
            spaces=" "*(hight-indexi-1)
            stars="*"*(indexj*2-1)
            print(spaces, end="")
            print(stars)
```

```python

## switch values of husband and wife and print both

husband='Amit'
wife="Poonam"
son="Amogh"
print(husband,' <--> ',wife,' <-->', son)
dummy=wife
wife=husband
husband=son
son=dummy

print(husband,' <--> ',wife,' <-->', son)
```
